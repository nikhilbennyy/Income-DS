```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
orginal = pd.read_csv(r'C:\Users\Nikhil\Desktop\Data Science\income(1).csv', na_values=" ?")
df = orginal.copy()
```

```python
df['SalStat'] = df['SalStat'].map({' less than or equal to 50,000':0, ' greater than 50,000':1})
df['SalStat'].value_counts()
```

```python
new_df = pd.get_dummies(df, drop_first=True)
new_df
```

```python
column_list = list(new_df.columns)
column_list
```

```python
features = list(set(column_list) - set(['SalStat']))
features
```

```python
y = new_df['SalStat'].values
y
```

```python
x = new_df[features].values
x
```

```python
train_x, test_x, train_y, test_y = train_test_split(x,y,test_size=0.1,random_state=0)
# test_y = list(test_y)
type(test_y)
```

```python
logistic = LogisticRegression(max_iter=10000)
```

```python
logistic.fit(train_x, train_y)
logistic.intercept_
```

```python
prediction = (logistic.predict(test_x))
type(prediction)
```

```python
confus = confusion_matrix(test_y, prediction)
confus
```

```python
print(classification_report(test_y, prediction))
```

```python
logistic.score(train_x, train_y)
```
